```python
import random

def bfs(n, grid, row_bot, col_bot, row_princess, col_princess):
    visited = [[False] * n for _ in range(n)]
    queue = [(row_bot, col_bot, [])]  # Estado: (fila, columna, lista de movimientos)

    while queue:
        row, col, movements = queue.pop(0)
        visited[row][col] = True

        if row == row_princess and col == col_princess:
            return movements

        # Definir las cuatro direcciones posibles (arriba, abajo, izquierda, derecha)
        directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
        moves = ["UP", "DOWN", "LEFT", "RIGHT"]

        for (dr, dc), move in zip(directions, moves):
            new_row, new_col = row + dr, col + dc
            if 0 <= new_row < n and 0 <= new_col < n and not visited[new_row][new_col]:
                new_movements = movements + [move]
                queue.append((new_row, new_col, new_movements))

def generate_random_coordinates(n):
    row_bot = random.randint(0, n - 1)
    col_bot = random.randint(0, n - 1)
    row_princess = random.randint(0, n - 1)
    col_princess = random.randint(0, n - 1)
    return row_bot, col_bot, row_princess, col_princess

def create_grid(n, row_bot, col_bot, row_princess, col_princess):
    grid = [["-" for _ in range(n)] for _ in range(n)]
    grid[row_bot][col_bot] = "m"
    grid[row_princess][col_princess] = "p"
    return grid

def nextMove(n, row_bot, col_bot, row_princess, col_princess):
    grid = create_grid(n, row_bot, col_bot, row_princess, col_princess)
    movements = bfs(n, grid, row_bot, col_bot, row_princess, col_princess)
    return movements

def step(grid, i):
    global row_bot
    global col_bot
    grid[row_bot][col_bot] = "-"
    print("[-----" + i + "-----]")
    if i == "DOWN":
        row_bot += 1
    elif i == "UP":
        row_bot -= 1
    elif i == "RIGHT":
        col_bot += 1
    elif i == "LEFT":
        col_bot -= 1
    grid[row_bot][col_bot] = "m"

    for row in grid:
        print("".join(row))
    print("\n")

# Solicitar al usuario el tamaño de la matriz
n = int(input("Ingrese el tamaño N de la matriz: "))

# Generar coordenadas aleatorias para el bot y la princesa
row_bot, col_bot, row_princess, col_princess = generate_random_coordinates(n)

# Calcular y ejecutar movimientos
movements = nextMove(n, row_bot, col_bot, row_princess, col_princess)
grid = create_grid(n, row_bot, col_bot, row_princess, col_princess)
for row in grid:
    print("".join(row))
print("\n")
for move in movements:
    step(grid, move)
```
